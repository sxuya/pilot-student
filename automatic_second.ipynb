```python
# 读取 Excel 文件
import pandas as pd

df = pd.read_excel('./assets/automatic/001-202002.xlsx', header = None)
df
```

```python
one_row = {}

one_row['姓名'] = df.iloc[3, 2]
one_row['ID'] = df.iloc[4, 6]
one_row['开始时间'] = df.iloc[1, 7].to_pydatetime()
one_row['结束时间'] = df.iloc[1, 9].to_pydatetime()
total_row = df.loc[df[1] == '总计']
one_row['网络'] = total_row.iloc[0, 3]
one_row['交通'] = total_row.iloc[0, 4]
one_row['住宿'] = total_row.iloc[0, 5]
one_row['餐饮'] = total_row.iloc[0, 6]
one_row['招待'] = total_row.iloc[0, 7]
one_row['其他'] = total_row.iloc[0, 8]
one_row['合计'] = total_row.iloc[0, 9]

print(one_row)
```

```python
one_row_df = pd.DataFrame([one_row])
# one_row_df = pd.DataFrame(one_row)  # this code is not worked
one_row_df
```
